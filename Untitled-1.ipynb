```python
import argparse
from dataclasses import dataclass
from typing import Dict, List, Tuple

import numpy as np

try:
    from tensorflow.keras.datasets import mnist
except Exception:
    mnist = None


def one_hot(y: np.ndarray, num_classes: int = 10) -> np.ndarray:
    out = np.zeros((y.shape[0], num_classes), dtype=np.float32)
    out[np.arange(y.shape[0]), y.astype(int)] = 1.0
    return out


def load_mnist() -> Tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray]:
    if mnist is None:
        raise ImportError("tensorflow.keras is required to load MNIST in this script.")
    (x_train, y_train), (x_test, y_test) = mnist.load_data()
    x_train = x_train.reshape(-1, 784).astype(np.float32) / 255.0
    x_test = x_test.reshape(-1, 784).astype(np.float32) / 255.0
    return x_train, y_train.astype(np.int64), x_test, y_test.astype(np.int64)


def relu(x: np.ndarray) -> np.ndarray:
    return np.maximum(0.0, x)


def relu_grad(x: np.ndarray) -> np.ndarray:
    return (x > 0).astype(np.float32)


def softmax(logits: np.ndarray) -> np.ndarray:
    shifted = logits - np.max(logits, axis=1, keepdims=True)
    exp = np.exp(shifted)
    return exp / np.sum(exp, axis=1, keepdims=True)


def cross_entropy(probs: np.ndarray, y_true: np.ndarray) -> float:
    eps = 1e-12
    return float(-np.sum(y_true * np.log(probs + eps)) / y_true.shape[0])


@dataclass
class LayerParams:
    W: np.ndarray
    b: np.ndarray
    vW: np.ndarray
    vb: np.ndarray


class NeuralNetwork:
    def __init__(
        self,
        layer_sizes: List[int],
        lr: float = 0.01,
        momentum: float = 0.9,
        l2: float = 1e-4,
        seed: int = 42,
    ):
        np.random.seed(seed)
        self.lr = lr
        self.momentum = momentum
        self.l2 = l2
        self.layers: List[LayerParams] = []

        for i in range(len(layer_sizes) - 1):
            fan_in, fan_out = layer_sizes[i], layer_sizes[i + 1]
            scale = np.sqrt(2.0 / fan_in) if i < len(layer_sizes) - 2 else np.sqrt(1.0 / fan_in)
            W = np.random.randn(fan_in, fan_out).astype(np.float32) * scale
            b = np.zeros((1, fan_out), dtype=np.float32)
            self.layers.append(LayerParams(W=W, b=b, vW=np.zeros_like(W), vb=np.zeros_like(b)))

    def forward(self, X: np.ndarray) -> Dict[str, List[np.ndarray]]:
        cache: Dict[str, List[np.ndarray]] = {"A": [X], "Z": []}
        A = X
        for i, layer in enumerate(self.layers):
            Z = A @ layer.W + layer.b
            cache["Z"].append(Z)
            A = relu(Z) if i < len(self.layers) - 1 else softmax(Z)
            cache["A"].append(A)
        return cache

    def compute_loss(self, probs: np.ndarray, y_onehot: np.ndarray) -> float:
        data_loss = cross_entropy(probs, y_onehot)
        reg_loss = 0.5 * self.l2 * sum(np.sum(layer.W * layer.W) for layer in self.layers)
        return data_loss + reg_loss

    def backward(self, cache: Dict[str, List[np.ndarray]], y_onehot: np.ndarray) -> List[Tuple[np.ndarray, np.ndarray]]:
        grads: List[Tuple[np.ndarray, np.ndarray]] = []
        m = y_onehot.shape[0]
        dA = (cache["A"][-1] - y_onehot) / m

        for i in reversed(range(len(self.layers))):
            A_prev = cache["A"][i]
            layer = self.layers[i]
            dW = A_prev.T @ dA + self.l2 * layer.W
            db = np.sum(dA, axis=0, keepdims=True)
            grads.insert(0, (dW.astype(np.float32), db.astype(np.float32)))

            if i > 0:
                dA = (dA @ layer.W.T) * relu_grad(cache["Z"][i - 1])

        return grads

    def step(self, grads: List[Tuple[np.ndarray, np.ndarray]]) -> None:
        for layer, (dW, db) in zip(self.layers, grads):
            layer.vW = self.momentum * layer.vW - self.lr * dW
            layer.vb = self.momentum * layer.vb - self.lr * db
            layer.W += layer.vW
            layer.b += layer.vb

    def predict(self, X: np.ndarray, batch_size: int = 1024) -> np.ndarray:
        preds = []
        for i in range(0, X.shape[0], batch_size):
            cache = self.forward(X[i : i + batch_size])
            preds.append(np.argmax(cache["A"][-1], axis=1))
        return np.concatenate(preds)

    def accuracy(self, X: np.ndarray, y: np.ndarray) -> float:
        return float(np.mean(self.predict(X) == y))

    def train(
        self,
        X_train: np.ndarray,
        y_train: np.ndarray,
        X_val: np.ndarray,
        y_val: np.ndarray,
        epochs: int = 10,
        batch_size: int = 64,
        verbose: bool = True,
    ) -> Dict[str, List[float]]:
        y_train_oh = one_hot(y_train, 10)
        y_val_oh = one_hot(y_val, 10)
        history = {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": []}
        n = X_train.shape[0]

        for epoch in range(1, epochs + 1):
            indices = np.random.permutation(n)
            X_train = X_train[indices]
            y_train_oh = y_train_oh[indices]
            y_train = y_train[indices]

            for start in range(0, n, batch_size):
                end = start + batch_size
                Xb = X_train[start:end]
                yb = y_train_oh[start:end]
                cache = self.forward(Xb)
                grads = self.backward(cache, yb)
                self.step(grads)

            train_cache = self.forward(X_train[: min(5000, n)])
            val_cache = self.forward(X_val)
            train_loss = self.compute_loss(train_cache["A"][-1], y_train_oh[: min(5000, n)])
            val_loss = self.compute_loss(val_cache["A"][-1], y_val_oh)
            train_acc = self.accuracy(X_train[: min(5000, n)], y_train[: min(5000, n)])
            val_acc = self.accuracy(X_val, y_val)

            history["train_loss"].append(train_loss)
            history["val_loss"].append(val_loss)
            history["train_acc"].append(train_acc)
            history["val_acc"].append(val_acc)

            if verbose:
                print(
                    f"Epoch {epoch:02d} | train_loss={train_loss:.4f} train_acc={train_acc:.4f} | "
                    f"val_loss={val_loss:.4f} val_acc={val_acc:.4f}"
                )

        return history


def train_val_split(X: np.ndarray, y: np.ndarray, val_ratio: float = 0.1, seed: int = 42):
    np.random.seed(seed)
    idx = np.random.permutation(len(X))
    split = int(len(X) * (1 - val_ratio))
    train_idx, val_idx = idx[:split], idx[split:]
    return X[train_idx], y[train_idx], X[val_idx], y[val_idx]


def run_experiment(
    hidden_sizes: List[int],
    lr: float,
    batch_size: int,
    epochs: int,
    momentum: float,
    l2: float,
) -> Dict[str, float]:
    X_train, y_train, X_test, y_test = load_mnist()
    X_train, y_train, X_val, y_val = train_val_split(X_train, y_train, val_ratio=0.1)
    layer_sizes = [784] + hidden_sizes + [10]
    model = NeuralNetwork(layer_sizes, lr=lr, momentum=momentum, l2=l2)
    model.train(X_train, y_train, X_val, y_val, epochs=epochs, batch_size=batch_size)
    test_acc = model.accuracy(X_test, y_test)
    print(f"Test accuracy: {test_acc:.4f}")
    return {"test_acc": test_acc}


def grid_search():
    configs = [
        {"hidden_sizes": [128], "lr": 0.01, "batch_size": 64, "epochs": 5, "momentum": 0.9, "l2": 1e-4},
        {"hidden_sizes": [256], "lr": 0.01, "batch_size": 64, "epochs": 5, "momentum": 0.9, "l2": 1e-4},
        {"hidden_sizes": [256, 128], "lr": 0.005, "batch_size": 64, "epochs": 5, "momentum": 0.9, "l2": 1e-4},
    ]

    best = None
    for cfg in configs:
        print(f"Running config: {cfg}")
        result = run_experiment(**cfg)
        score = result["test_acc"]
        if best is None or score > best["score"]:
            best = {"score": score, "cfg": cfg}
    print(f"Best: {best}")


def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("--epochs", type=int, default=10)
    parser.add_argument("--batch-size", type=int, default=64)
    parser.add_argument("--lr", type=float, default=0.01)
    parser.add_argument("--momentum", type=float, default=0.9)
    parser.add_argument("--l2", type=float, default=1e-4)
    parser.add_argument("--hidden-sizes", type=str, default="128,64")
    parser.add_argument("--grid-search", action="store_true")
    args = parser.parse_args()

    if args.grid_search:
        grid_search()
        return

    hidden_sizes = [int(x) for x in args.hidden_sizes.split(",") if x.strip()]
    run_experiment(
        hidden_sizes=hidden_sizes,
        lr=args.lr,
        batch_size=args.batch_size,
        epochs=args.epochs,
        momentum=args.momentum,
        l2=args.l2,
    )


# Notebook-safe: do not auto-run argparse in Jupyter.
# Run manually if needed:
run_experiment(hidden_sizes=[128, 64], lr=0.01, batch_size=64, epochs=10, momentum=0.9, l2=1e-4)
```
